## 🔹 Part 1: Using PIL (Pillow)

PIL loads images as **PIL Image objects**, not NumPy arrays.


```python

# TODO 1:
# 1. Open the image "images/test_image.jpg" using PIL
# 2. Print the image type, size, and mode
# 3. Display the image

from PIL import Image

# YOUR CODE HERE
img = Image.open("images/test_image.jpg")

print("Type:", type(img))
print("Size:", img.size)   # (width, height)
print("Mode:", img.mode)   # RGB, L, RGBA

img.show()
```


### ✨ Task 1.1: PIL → NumPy Conversion


```python

# TODO 2:
# Convert the PIL image to a NumPy array
# Print the array shape and data type

import numpy as np

# YOUR CODE HERE
# Convert PIL image to NumPy array
img_np = np.array(img)

# Print shape and data type
print("Shape:", img_np.shape)
print("Data type:", img_np.dtype)
```


## 🔹 Part 2: Matplotlib Image Reading

Matplotlib reads images directly as NumPy arrays.


```python

# TODO 3:
# 1. Read the image using matplotlib.image.imread
# 2. Print its shape
# 3. Display it using plt.imshow

import matplotlib.image as mpimg
import matplotlib.pyplot as plt

# YOUR CODE HERE
img_mp = mpimg.imread("images/test_image.jpg")

print("Shape:", img_mp.shape)

plt.imshow(img_mp)
plt.axis("off")  
plt.show()
```


### ❓ Concept Question
**Q:** Why does Matplotlib treat images as NumPy arrays?
Write your answer below.


#### My Answer:
Because images are naturally stored as grids of pixel values, and NumPy arrays are an efficient way to represent and process this numerical data for visualization and computation.


## 🔹 Part 3: scikit-image

scikit-image works with well-defined data ranges.


```python

# TODO 4:
# 1. Read the image using skimage.io.imread
# 2. Convert it to float using img_as_float
# 3. Print min and max pixel values
# 4. Display the image

from skimage import io, img_as_float
import matplotlib.pyplot as plt

# YOUR CODE HERE
img_sk = io.imread("images/test_image.jpg")

# Convert image to float with proper normalization
img_float = img_as_float(img_sk)

# Print minimum and maximum pixel values
print("Min pixel value:", img_float.min())
print("Max pixel value:", img_float.max())

plt.imshow(img_float)
plt.axis("off")
plt.show()
```


### ✨ Task 3.1: Data Type Exploration

**Q:** Why is using `astype(float)` discouraged compared to `img_as_float`?
Explain briefly.


#### My Answer:
Because astype(float) only changes the data type without scaling pixel values, while img_as_float correctly converts and normalizes the image to the expected range [0, 1], preventing incorrect intensity values.


## 🔹 Part 4: OpenCV

⚠️ OpenCV loads color images in **BGR** format.


```python

# TODO 5:
# 1. Read the image in grayscale and color using OpenCV
# 2. Convert the color image from BGR to RGB
# 3. Display the RGB image using matplotlib

import cv2
import matplotlib.pyplot as plt

# YOUR CODE HERE
# Read image in grayscale
img_gray = cv2.imread("images/test_image.jpg", cv2.IMREAD_GRAYSCALE)

# Read image in color (BGR format)
img_bgr = cv2.imread("images/test_image.jpg", cv2.IMREAD_COLOR)

# Convert BGR to RGB
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# Display the RGB image using matplotlib
plt.imshow(img_rgb)
plt.axis("off")
plt.show()
```


### ✨ Task 4.1: Image Processing

Apply **Canny edge detection** on the grayscale image and display the result.


```python

# TODO 6:
# Perform edge detection using cv2.Canny

# YOUR CODE HERE
# Canny edge detection
edges = cv2.Canny(img_gray, threshold1=100, threshold2=200)

plt.imshow(edges, cmap="gray")
plt.axis("off")
plt.show()
```


### ⭐ Bonus Task B: Reading Multiple Images

Read all images from a folder and display them one by one.


```python

# BONUS (Optional)

import glob
import cv2
import matplotlib.pyplot as plt


path = "images/test_images/aeroplane/*.*"

# Loop through all image files
for file in glob.glob(path):
    # Read image in color (BGR)
    img_bgr = cv2.imread(file)
    
    # Convert BGR to RGB
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    
    # Display the image
    plt.imshow(img_rgb)
    plt.title(file.split("/")[-1])  
    plt.axis("off")
    plt.show()
```
