# Part 1: Histogram Equalization

Histogram equalization improves contrast in images by redistributing the intensity values to make the histogram as flat as possible.

```python
import cv2
import matplotlib.pyplot as plt

# Read a grayscale image
image = cv2.imread('/content/drive/MyDrive/DIP/ai-image.jpg', cv2.IMREAD_GRAYSCALE)

# Apply histogram equalization
equalized_image = cv2.equalizeHist(image)

# Create figure
plt.figure(figsize=(10, 5))

# ---- Original Image ----
plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Histogram of Original Image
# ravel() flattens the 2D image matrix into a 1D array
# so we can compute its histogram
plt.subplot(2, 2, 2)
plt.hist(image.ravel(), 256, [0, 256])
plt.title('Histogram (Original)')

# ---- Equalized Image ----
plt.subplot(2, 2, 3)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')

# Histogram of Equalized Image
plt.subplot(2, 2, 4)
plt.hist(equalized_image.ravel(), 256, [0, 256])
plt.title('Histogram (Equalized)')

plt.tight_layout()
plt.show()
```

## Explanation of `ravel()`

The `ravel()` function from NumPy flattens a multi-dimensional array into a one-dimensional array. In the case of images, it is used to convert the 2D image matrix into a 1D array for easy histogram plotting.

## 1. Apply Histogram Equalization and Observe the Effect

### Using OpenCV

```python
import cv2
import matplotlib.pyplot as plt

# Read grayscale image
image = cv2.imread('/content/drive/MyDrive/DIP/ai-image.jpg', cv2.IMREAD_GRAYSCALE)

# Apply histogram equalization
equalized_image = cv2.equalizeHist(image)

# Display images
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(equalized_image, cmap='gray')
plt.title("Equalized Image")
plt.axis('off')

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.hist(image.ravel(), 256, [0,256])
plt.title("Histogram (Original)")

plt.subplot(1,2,2)
plt.hist(equalized_image.ravel(), 256, [0,256])
plt.title("Histogram (Equalized)")

plt.show()
```

## 3. Why Histogram Equalization Improves Contrast in Poor Lighting

### In Poorly Lit Images

- Pixel values are concentrated in a small intensity range (often dark).  
- There is little difference between neighboring pixel values.  
- The image looks dull or flat.  

---

###  What Histogram Equalization Does

- It redistributes pixel intensity values.  
- It stretches frequently occurring intensity ranges.  
- It spreads the histogram across the full range (0–255).  
- It increases the dynamic range of pixel intensities.  

---

###  Result

- Dark pixels become darker where needed.  
- Bright pixels become brighter.  
- The difference between regions increases.  
- Human visual perception improves because contrast increases.  

---

### In Simple Terms

Histogram equalization spreads out the most frequent intensity values, allowing areas of lower contrast to gain higher contrast.

# Part 2: Histogram Equalization on Color Images

## Step 2: Applying Histogram Equalization to Color Channels Separately

When working with color images, histogram equalization can be applied to individual color channels (Red, Green, Blue) separately.

```python
import cv2
import matplotlib.pyplot as plt

# Read color image
color_image = cv2.imread('/content/drive/MyDrive/DIP/ai-image.jpg')

# Split into B, G, R channels
b, g, r = cv2.split(color_image)

# Apply histogram equalization to each channel
equalized_b = cv2.equalizeHist(b)
equalized_g = cv2.equalizeHist(g)
equalized_r = cv2.equalizeHist(r)

# Merge channels back
equalized_color_image = cv2.merge([equalized_b, equalized_g, equalized_r])

# Display images
plt.figure(figsize=(10,5))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(color_image, cv2.COLOR_BGR2RGB))
plt.title('Original Color Image')
plt.axis('off')

# Equalized Image
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(equalized_color_image, cv2.COLOR_BGR2RGB))
plt.title('Equalized Color Image (Per Channel)')
plt.axis('off')

plt.show()
```

# Part 3: Histogram Matching (Specification)

Histogram matching transforms the histogram of an input image to match a target histogram from another image. This is useful for matching brightness and contrast between two images.

---

## Step 3: Applying Histogram Matching

### Using `skimage` (Recommended Method)

```python
import cv2
import numpy as np
from skimage import exposure
import matplotlib.pyplot as plt

# Read two grayscale images
image_source = cv2.imread('/content/drive/MyDrive/DIP/ai-image.jpg', cv2.IMREAD_GRAYSCALE)
image_target = cv2.imread('/content/drive/MyDrive/DIP/blue_rose.jpg', cv2.IMREAD_GRAYSCALE)

# Perform histogram matching
matched_image = exposure.match_histograms(image_source, image_target)

# Display results
plt.figure(figsize=(15,5))

# Source Image
plt.subplot(3,3,1)
plt.imshow(image_source, cmap='gray')
plt.title('Source Image')
plt.axis('off')

plt.subplot(3,3,2)
plt.hist(image_source.ravel(), 256, [0,256])
plt.title('Histogram (Source)')

# Target Image
plt.subplot(3,3,4)
plt.imshow(image_target, cmap='gray')
plt.title('Target Image')
plt.axis('off')

plt.subplot(3,3,5)
plt.hist(image_target.ravel(), 256, [0,256])
plt.title('Histogram (Target)')

# Matched Image
plt.subplot(3,3,7)
plt.imshow(matched_image, cmap='gray')
plt.title('Matched Image')
plt.axis('off')

plt.subplot(3,3,8)
plt.hist(matched_image.ravel(), 256, [0,256])
plt.title('Histogram (Matched)')

plt.show()
```

## Explanation of `exposure.match_histograms()`

The `exposure.match_histograms()` function from **skimage** adjusts the pixel intensities of a source image so that its histogram matches that of a target image. It is useful for ensuring visual consistency between images in terms of contrast and brightness.

## Task 3

### 1. Choose Two Images (Source and Target)  
- **Source image**: The image you want to modify.  
- **Target image**: The image whose histogram will be used as reference.

---

### 2. Apply Histogram Matching

```python
import cv2
from skimage import exposure
import matplotlib.pyplot as plt

# Read images in grayscale
source = cv2.imread('/content/drive/MyDrive/DIP/ai-image.jpg', cv2.IMREAD_GRAYSCALE)
target = cv2.imread('/content/drive/MyDrive/DIP/blue_rose.jpg', cv2.IMREAD_GRAYSCALE)

# Apply histogram matching
matched = exposure.match_histograms(source, target)

# Display images
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(source, cmap='gray')
plt.title("Source Image")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(target, cmap='gray')
plt.title("Target Image")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(matched, cmap='gray')
plt.title("Matched Image")
plt.axis('off')

plt.show()
```

## What Happens After Histogram Matching?

- The intensity distribution of the source image changes.  
- The histogram of the source becomes similar to the target histogram.  
- Brightness and contrast adjust according to the target image.  

---

##  3. How Histogram Matching Affects Visual Appearance

###  Visual Changes:

- The overall brightness of the source image may increase or decrease.  
- Contrast changes to resemble the target image.  
- Dark images can become brighter if the target is bright.  
- Bright images can become darker if the target is dark.  
- The image appearance becomes similar to the target image.  

## Explanation

Histogram matching works by:

- Transforming pixel intensity values.  
- Matching the cumulative distribution function (CDF) of the source image to that of the target image.  
- Adjusting intensity values so both histograms look similar.  

In this way, the source image is modified to follow the same intensity distribution as the target image.
