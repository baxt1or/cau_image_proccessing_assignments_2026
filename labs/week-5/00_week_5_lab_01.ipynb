```python
import numpy as np 
import cv2 
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("./images/img1.jpeg")[:, :, 0]
img_eq = cv2.equalizeHist(img)

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis("off")

plt.subplot(2, 2, 2)
plt.hist(img.ravel(), bins=80)
plt.title("Original Histogram")

plt.subplot(2, 2, 3)
plt.imshow(img_eq, cmap='gray')
plt.title("Equalized Image")
plt.axis("off")

plt.subplot(2, 2, 4)
plt.hist(img_eq.ravel(), bins=80)
plt.title("Equalized Histogram")

plt.tight_layout()
plt.show()
```

```python
def task2(img):
    img_eq = cv2.equalizeHist(img)

    plt.figure(figsize=(12, 8))

    plt.subplot(2, 2, 1)
    plt.imshow(img, cmap='gray')
    plt.title("Original Image")
    plt.axis("off")

    plt.subplot(2, 2, 2)
    plt.hist(img.ravel(), bins=80)
    plt.title("Original Histogram")

    plt.subplot(2, 2, 3)
    plt.imshow(img_eq, cmap='gray')
    plt.title("Equalized Image")
    plt.axis("off")

    plt.subplot(2, 2, 4)
    plt.hist(img_eq.ravel(), bins=80)
    plt.title("Equalized Histogram")

    plt.tight_layout()
    plt.show()



img2 = cv2.imread("./images/img1.jpeg")

# B dimention
task2(img2[:, :, 1])
```

```python
task2(img2[:, :, 2])
```

```python
img3 = cv2.imread("./images/kuala.jpg")[:, :, ::-1]

merged_channels = np.stack(np.array([cv2.equalizeHist(img3[:, :, i]) for i in range(3)]), axis=-1)

plt.figure(figsize=(12, 8))

plt.subplot(1, 2, 1)
plt.imshow(merged_channels)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.hist(merged_channels.reshape(-1), bins=80)
plt.title("Original Histogram")
```

```python
import glob
from skimage import exposure 

files = glob.glob("./images/*")

images = [cv2.equalizeHist(cv2.imread(img)[:, :, 0]) for img in files]

mt_img = exposure.match_histograms(images[0], images[1])

plt.imshow(mt_img)
```
